```python
with open('data/test.txt', 'r', encoding='utf-8') as infile:
    lines = infile.readlines()
    lines = [line.strip() for line in lines]
print(f'Total lines: {len(lines)}')
```

```python
from predict import predict_sentence
from vietnamese_utils import remove_vietnamese_tone


def evaluate(sentence):
    no_tone_sentence = remove_vietnamese_tone(sentence)
    predicted_sentence = predict_sentence(no_tone_sentence)

    original_words = sentence.split()
    predicted_words = predicted_sentence.split()

    correct = 0
    for w1, w2 in zip(original_words, predicted_words):
        if w1 == w2:
            correct += 1
    return correct, len(original_words), no_tone_sentence, sentence, predicted_sentence
```

```python
from utils import get_display

accuracy_display = get_display('Accuracy: 0')
no_tone_sentence_display = get_display('no_tone_sentence')
target_sentence_display = get_display('target_sentence')
output_sentence_display = get_display('output_sentence')

total_correct = 0
total_words = 0
for i, line in enumerate(lines, 1):
    try:
        correct, total, no_tone_sentence, target_sentence, output_sentence = evaluate(line)
        total_correct += correct
        total_words += total
        accuracy_display.update(f'{i}. Accuracy: {total_correct / total_words:0.4f}')
        no_tone_sentence_display.update(f'Input : {no_tone_sentence}')
        target_sentence_display.update(f'Target: {target_sentence}')
        output_sentence_display.update(f'Output: {output_sentence}')
    except Exception as ex:
        print(line)
        print(ex)
```
