```python
from pprint import pprint
```

# Load data

```python
from zipfile import ZipFile

with ZipFile('data/vietnamese_tone_prediction.zip', 'r') as inzip:
    data = inzip.read('train.txt').decode('utf-8').split('\n')

print(f'Có {len(data):_} câu')
pprint(data[:5])
```

# Filter data

```python
import random

random.seed()

data = [d for d in data if 50 < len(d) < 140]
random.shuffle(data)

print(f'Có {len(data):_} câu')
pprint(data[:5])
```

# Chia tập data

```python
ratio = 0.9
split_point = int(len(data) * ratio)

train_set = data[:split_point]
test_set = data[split_point:]

print(f'Tập train: {len(train_set)} câu')
print(f'Tập test: {len(test_set)} câu')
```

# Ghi file train.txt và test.txt

```python
with open('data/train.txt', 'w', encoding='utf-8') as outfile:
    train_data = '\n'.join(train_set)
    outfile.write(train_data)
    print('Đã ghi train.txt')
    
with open('data/test.txt', 'w', encoding='utf-8') as outfile:
    test_data = '\n'.join(test_set)
    outfile.write(test_data)
    print('Đã ghi text.txt')
```
